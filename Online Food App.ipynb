```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Feedback",axis=1)
y = df["Feedback"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,
                                                y,
                                                test_size=0.2,
                                                random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#model building
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR
```

```python
#model fitting

LR.fit(X_train,y_train)
```

```python
#model predictions

y_pred = LR.predict(X_test)

print(y_test.values[:5])
print(y_pred[:5])
```

```python
from sklearn.metrics import precision_score,accuracy_score,f1_score,recall_score,roc_auc_score,roc_curve
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report
```

```python
acc_LR = round(accuracy_score(y_test,y_pred)*100,2)
pre_LR = precision_score(y_test,y_pred)
rec_LR = recall_score(y_test,y_pred)
f1_LR = f1_score(y_test,y_pred)

print("Accuracy :",acc_LR)
print("Precision :",pre_LR)
print("Recall :",rec_LR)
print("F1 Score :",f1_LR)
```

```python
cmt = confusion_matrix(y_test,y_pred)
cmt

disp = ConfusionMatrixDisplay(confusion_matrix=cmt,display_labels=[False,True])
disp.plot()
plt.grid(False)
plt.show()
```

```python
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

print("True Negative :",tn)
print("False positive :",fp)
print("False Negative :",fn)
print("True Positive :",tp)
```

```python
y_pred_prob = LR.predict_proba(X_test)[:,1]
fpr,tpr,threshold = roc_curve(y_test,y_pred_prob)
plt.plot(fpr,tpr)
plt.show()
```

```python
acc_LR = round(accuracy_score(y_test,y_pred)*100,2)
pre_LR = precision_score(y_test,y_pred)
rec_LR = recall_score(y_test,y_pred)
f1_LR = f1_score(y_test,y_pred)

print("Accuracy :",acc_LR)
print("Precision :",pre_LR)
print("Recall :",rec_LR)
print("F1 Score :",f1_LR)


cmt = confusion_matrix(y_test,y_pred)
cmt

disp = ConfusionMatrixDisplay(confusion_matrix=cmt,display_labels=[False,True])
disp.plot()
plt.grid(False)
plt.show()


tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

print("True Negative :",tn)
print("False positive :",fp)
print("False Negative :",fn)
print("True Positive :",tp)


y_pred_prob = LR.predict_proba(X_test)[:,1]
fpr,tpr,threshold = roc_curve(y_test,y_pred_prob)
plt.plot(fpr,tpr)
plt.show()
```

```python
# TN : Actual is NEG , Model is also predicting NEG

#TP : Axtual is POS ,MOdel is also predicting POS

#FN : Actual is POS , BUt model is predicting NEG

#FP : Actual is NEG , But model is predicting POS
```

```python
#Accuracy : tp+tn/tp+tn+fp+fn

#Precision : Precision= 

# True Positives/True Positives+False Positives
 

# In other words, precision answers the question: 
#     "Of all the instances predicted as positive, how many were actually positive?"

#Recall : Recall= 

# True Positives/True Positives+False Negatives

 

# In other words, recall answers the question: 
#     "Of all the actual positive instances, how many did the model correctly predict as positive?"

#f1_score = Harmonic mean of precision and recall
```
